```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib as plt
```

```python
df_parquet = pd.read_parquet('../data/reservas_hoteles.parquet')
df_parquet.head()
```

Compruebo si los hoteles de la competencia son 10

```python
hoteles_competencia = df_parquet[(df_parquet["competencia"] == True)]
num_ids_diferentes = hoteles_competencia["id_hotel"].nunique()
num_ids_diferentes
```

```python
df_parquet.info()
```

Sacar valores unicos de la comlumna "Nombre hotel"

```python
df_hoteles = df_parquet["nombre_hotel"].unique()
df_hoteles
```

```python
df_parquet['nombre_hotel'] = df_parquet['nombre_hotel'].replace('', pd.NA)
```

```python
df_hoteles = df_parquet["nombre_hotel"].unique()
df_hoteles
```

Asignar valores unicos a cada hotel

```python
df_parquet["id_hotel_final"] = ["ID" + str(i) for i in pd.factorize(df_parquet["nombre_hotel"])[0]]
df_parquet.head()
```

Sustituimos los ID creados para el valor NaN por NaN tambien para no tener confuciones.

```python
df_parquet['id_hotel_final'] = df_parquet['id_hotel_final'].replace('ID-1',pd.NA)
```

```python
df_parquet
```

Rellenamos los valores NaN de las columnas inicio y final_estancia con los indicados.

```python
df_parquet.fillna({'inicio_estancia': '2025-03-01', 'final_estancia': '2025-03-02'}, inplace=True)
df_parquet.info()
```

Ahora utilizamos una función para rellenar todas las casillas de la columna ciudad que contienen un valor string vacio '' o None por Madrid

```python
df_parquet['ciudad'] = df_parquet['ciudad'].apply(lambda x: 'Madrid' if x == '' or x is None else x)
df_parquet.head()
```

Comprobamos duplicados en id_cliente y los contabilizamos.

```python
df_parquet['id_cliente'].duplicated().value_counts()
```

```python
df_id_duplicados = df_parquet[df_parquet['id_cliente'].duplicated()]
df_id_duplicados.head()
```

Comprobamos si estos registos de id duplicandos son correspondientes también a diferentes peronas y mails y así es. 

```python
duplicados_muestra = df_parquet[df_parquet['id_cliente'] == '346f2a77-33f4-4b2d-82f3-e8501e21b28d']
duplicados_muestra
```

Utilizamos los mail de los clientes para crear ID únicos para cada uno de ellos ya que como hemos comprobado anteriormente los id de cliente no son del todo precisos.

```python
df_parquet["id_cliente_final"] = ["ID" + str(i) for i in pd.factorize(df_parquet["mail"])[0]]
df_parquet.head()
```

Rellenamos los valores nulos de la columna estrellas de los hoteles que si tenemos valoraciones con la media para cada hotel.

```python
df_parquet['estrellas'] = df_parquet.groupby('nombre_hotel')['estrellas'].transform('mean')
```

```python
df_parquet['estrellas'] = df_parquet['estrellas'].round(1)
df_parquet.head()
```

```python
media_precio = df_parquet.groupby('nombre_hotel')['precio_noche'].transform('mean')
```

```python
df_parquet['precio_noche'] = df_parquet['precio_noche'].fillna(media_precio)
```
